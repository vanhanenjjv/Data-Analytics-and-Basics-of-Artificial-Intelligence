```python
import numpy as np
import pandas as pd
```

```python
emps = pd.read_csv("data/employees.csv")
dpts = pd.read_csv("data/departments.csv")

employees = emps.merge(dpts, how="inner", on="dep")
employees.drop("image", inplace=True, axis=1)

employees.style.set_caption("Employees").hide_index()
```

```python
men = employees.query("gender == 0")

men.style.set_caption("Men").hide_index()
```

```python
women = employees.query("gender == 1")

women.style.set_caption("Women").hide_index()
```

```python
percentages = pd.DataFrame([
    ["Men", len(men) / len(employees) * 100],
    ["Women", len(women) / len(employees) * 100]
], columns=["Gender", "Percentage"])

percentages.style.format({ "Percentage" : "{:.1f}%" }).hide_index()
```

```python
salaries = pd.DataFrame([
    ["Max", employees["salary"].max()],
    ["Avg", employees["salary"].mean()],
    ["Min", employees["salary"].min()]
], columns=["Type", "Amount"])

salaries.style.format({ "Amount": "{:.0f}" }).set_caption("All employees' wages").hide_index()
```

```python
from datetime import date

today = date.today()

employees['age'] = employees['bdate'].apply(lambda bdate: round(np.floor((today - date.fromisoformat(bdate)).days / 365)))
employees
```

```python
def age_group_of (age):
    if age > 70: return 70
    if age < 15: return 15

    for n in range(70, 15, -5):
        m = age - n
        if m <= 0 and m > -5:
            return n

employees['age_group'] = employees['age'].apply(age_group_of)
employees
```
